#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
# your answer here

aggregation_years = employee.groupby('Department')['Years'].mean()
aggregation_salary = employee.groupby('Department')['Salary'].mean()

def min_value(int):
    result = int.min()
    return result

def max_value(int):
    result = int.max()
    return result

def mean_value(int):
    result = int.mean()
    return result

print(min_value(aggregation_years))
print(min_value(aggregation_salary))
print(max_value(aggregation_years))
print(max_value(aggregation_salary))
print(mean_value(aggregation_years))
print(mean_value(aggregation_salary))
```
